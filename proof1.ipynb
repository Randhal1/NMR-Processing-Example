```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data=pd.DataFrame(columns=['f1','r1','f2','r2','f3','r3','f4','r4','f5','r5','f6','r6'])
```

```python
data[['f1','r1']]=pd.read_csv('001.txt',' ')
data[['f2','r2']]=pd.read_csv('002.txt',' ')
data[['f3','r3']]=pd.read_csv('003.txt',' ')
data[['f4','r4']]=pd.read_csv('004.txt',' ')
data[['f5','r5']]=pd.read_csv('005.txt',' ')
data[['f6','r6']]=pd.read_csv('006.txt',' ')
data
```

```python
plt.figure(figsize=[15,15])
plt.plot(data['f1'],data['r1'],label='w1')
#plt.plot(data['f2'],data['r2'],label='w2')
#plt.plot(data['f3'],data['r3'],label='w3')
plt.plot(data['f4'],data['r4'],label='g1')
#plt.plot(data['f5'],data['r5'],label='g2')
#plt.plot(data['f6'],data['r6'],label='g3')
plt.grid()
plt.legend(loc='best')
```

```python
data['prom_water']=(data['r1']+data['r2']+data['r3'])/3
```

```python
data['prom_eco']=(data['r4']+data['r5']+data['r6'])/3
```

```python
plt.figure(figsize=[15,5])
plt.plot(data['f4'],data['prom_water'],label='Espectro del agua')
plt.plot(data['f4'],data['prom_eco'],label='Espectro de fondo')
plt.grid()
plt.legend()
plt.xlabel('Frecencia')
plt.ylabel('RMS')
```

```python
data['water']=data['prom_water']-data['prom_eco']
```

```python
plt.figure(figsize=[15,5])
plt.plot(data['f4'],data['water'],label='water')
plt.grid()
plt.legend()
plt.xlabel('Frecencia')
plt.ylabel('RMS')
```

```python
from scipy import fft
```

```python
u=fft(data['water'])
```

```python
plt.figure(figsize=[15,5])
plt.plot(data['f4'],np.log(u),label='water')
plt.grid()
plt.legend()
plt.xlabel('Frecuencia')
plt.ylabel('log_fft_RMS')
```
